```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# a)

college = pd.read_csv('college.csv')
college.head()
```

```python
# b)

college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
college=college3
college.head()
```

```python
# c)

college.describe()
```

```python
# d)

pd.plotting.scatter_matrix(college[['Top10perc','Apps','Enroll']],alpha=0.5)
```

```python
# e)

college.boxplot(column='Outstate', by='Private')
```

```python
# f)

college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
# f)

college.boxplot(column='Outstate',by='Elite')
```

```python
# g)

college.hist(['Apps', 'PhD', 'Room.Board', 'Books'],grid=False,edgecolor='black',rwidth=0.8)
```

### Further data exploration

```python
college_corr = pd.get_dummies(college).corr().round(2)
college_corr
```

```python
plt.figure(figsize=(10,10))
corr_matrix = college_corr[college_corr.columns].astype(float)
sns.heatmap(corr_matrix, cmap='crest',annot=True,annot_kws={'size':6},linewidths=.2)
plt.title('The correlation coefficients')
plt.show()
```

```python
sns.regplot(data=college, x='Apps',y='Expend',logx=True)
college['Apps'].corr(college['Expend'])
```

```python
sns.regplot(data=college, x='Apps',y='PhD',logx=True)
college['Apps'].corr(college['PhD'])
```

### Further data exploration<br>
#### Correlation coefficients<br>
The goal of data exploration is to identify and analyze hidden patterns and relationships in College dataset. The first step is to calculate pairwise correllation of all columns. Because the missing values can potentially lead to biased results and invalid conclusions the overall missing rate should be calculated.

```python
college.isna().sum().tolist()
```

Pairwise correlation can be computed in the usual way because there are no missing values in the dataset. <br><br>
Coefficient +1 means a positive correlation<br>
Coefficient -1 means a negative correlation<br>
Coefficient 0 means there is no correlation.

```python
# creates a new DataFrame where categorical columns transforms into numerical in order to be able to compute corellation coefficient
# computes pairwise correlation of columns
college_corr = pd.get_dummies(college).corr().round(2)
college_corr.head()
```

#### Visual data exploration
The correlation matrix shows how pairs of values are related to each other.

```python
plt.figure(figsize=(10,10))
corr_matrix = college_corr[college_corr.columns].astype(float)
sns.heatmap(corr_matrix, cmap='crest',annot=True,annot_kws={'size':6},linewidths=.2)
plt.title('The correlation coefficients')
plt.show()
```

#### Interpretation of the correlation matrix. <br>
The number of applications received (**Apps**) is positively correlated with the number of applications accepted (**Accept**), the number of new students enrolled (**Enroll**), the number of full-time undergraduates (**F.Undergrad**) and is negatively correlated with the percent of alumni who donate (**perc.alumni**), private (**Private_Yes**) and non-elite institutions (**Elite_No**).<br>

The percent of faculty with Ph.D.s (**PhD**) is positively correlated with the number of new students from top 10 (**Top10perc**), the number of new students from top 25 (**Top25perc**) and is negatively correlated with estimated personal spending (**Personal**), student/faculty ratio (**S.F.Ratio**) and non-elite institutions (**Elite_No**).<br>

Elite institutions (**Elite_Yes**) are positively correlated with the number of new students from top 10 (**Top10perc**), instructional expenditure per student (**Expend**) and are negatively correlated with student/faculty ratio (**S.F.Ratio**), while non-elite institutions are negatively correlated with the number of applications received (**Apps**), the number of new students from top 10 (**Top10perc**) and instructional expenditure per student (**Expend**).<br>

Private institutions (**Private_Yes**) are positively correlated with out-of-state tuition (**Outstate**) and are negatively correlated with the number of applications received (**Apps**), the number of applications accepted (**Accept**), the number of new students enrolled (**Enroll**), the number of full-time undergraduates (**F.Undergrad**) and student/faculty ratio (**S.F.Ratio**).<br> Non-private institutions (**Private_No**), on the contrary, are positively correlated with the number of applications received (**Apps**), the number of applications accepted (**Accept**), the number of new students enrolled (**Enroll**), the number of full-time undergraduates (**F.Undergrad**) and is negatively correlated with out-of-state tuition (**Outstate**).<br>

Instructional expenditure per student (**Expend**) is positively correlated with the number of new students from top 10 (**Top10perc**), the number of new students from top 25 (**Top25perc**), out-of-state tuition (**Outstate**), elite institutions (**Elite_Yes**) and is negatively correlated with student/faculty ratio (**S.F.Ratio**) and non-elite institutions (**Elite_No**).<br>
